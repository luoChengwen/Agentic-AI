````python
def generate_responses(mesages: List[Dict]):
    response = completion(
        model = 'gpt-4o',
        messages=messages,
        max_tokens=1024
    )
    return response.choices[0].message['content']
````

````python
messages = [
    {'role': 'system', 'content': 'You are an expert of python code writing.'}
]

print('Please write a function you want')
function_description = input().strip()

messages.append(
    {'role': 'user',
        'content': f"wirte a function in python - {function_description}. Output it in '''python code block''' "
    }
)

init_function = generate_responses(messages)
print(init_function)
````
